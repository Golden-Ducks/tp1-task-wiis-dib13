```python
import csv

def clean(text):
    text = text.lower()
    ponctuation = "!?.,'\"-;:()"
    for c in ponctuation:
        text = text.replace(c, " ")
    return text.split()

def vect(doc, vocab):
    v = []
    for mot in vocab:
        if mot in doc:
            v.append(1)
        else:
            v.append(0)
    return v

fichier_fake = "/kaggle/input/datasets/emineyetm/fake-news-detection-datasets/News _dataset/Fake.csv"
fichier_true = "/kaggle/input/datasets/emineyetm/fake-news-detection-datasets/News _dataset/True.csv"

docs = []
labels = []

with open(fichier_fake, 'r', encoding='utf-8') as f:
    reader = csv.reader(f)
    next(reader)
    for i, row in enumerate(reader):
        if i >= 5: break
        docs.append(clean(" ".join(row)))
        labels.append(0)  # 0 = fake


with open(fichier_true, 'r', encoding='utf-8') as f:
    reader = csv.reader(f)
    next(reader)
    for i, row in enumerate(reader):
        if i >= 5: break
        docs.append(clean(" ".join(row)))
        labels.append(1)  # 1 = True

print("Nombre de docs:", len(docs))
print("Labels:", labels)
print()

vocab = []
for doc in docs:
    for mot in doc:
        if mot not in vocab:
            vocab.append(mot)

print("Taille vocabulaire:", len(vocab))
print("10 premiers mots:", vocab[:10])
print()

vecteurs = []
for doc in docs:
    vecteurs.append(vect(doc, vocab))

print("3 premiers vecteurs:")
for i in range(3):
    print(f"Doc {i} (label {labels[i]}): {vecteurs[i][:15]}...")
```
